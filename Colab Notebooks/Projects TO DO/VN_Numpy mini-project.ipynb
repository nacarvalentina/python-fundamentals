Use the code cell below to work your code.

```python
from pandas.core.indexing import convert_from_missing_indexer_tuple
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/lilaceri/Working-with-data-/main/Data%20Sets%20for%20code%20divisio/exams.csv"
df = pd.read_csv(url)

#using functions
math=np.array(df["math_score"],np.int64)
reading=np.array(df["reading_score"],np.int64)
writing=np.array(df["writing_score"],np.int64)

def calc_av(a,b,c):
  average= (a+b+c)/3 
  df["Average_score"]=average.tolist()   
  return  df
Av_score=calc_av(math,reading,writing)
Av_score 


def filtered_df_average(df,col,criteria):
  filtered_df=df[df[col]==criteria]  
  return filtered_df


average_Score_female=filtered_df_average(Av_score,"gender","female").mean()
average_Score_male=filtered_df_average(Av_score,"gender","male").mean()
average_Score_female
average_Score_male

filtered_f=filtered_df_average(Av_score,"gender","female")
filtered_m=filtered_df_average(Av_score,"gender","female")

def correlation(df1,col1,df2,col2):
  coef=np.corrcoef(df1[col1],df2[col2]) 
  return coef

coefficient1= correlation(df,"reading_score",df,"math_score")
print("Coefficient1: ",coefficient1) #instead of printing, you can just do that

coefficient2= correlation(df,"reading_score",df,"writing_score")
print("Coefficient2: ",coefficient2) 

coefficient_F= correlation(filtered_f,"reading_score",filtered_f,"writing_score")
print("CoefficientF: ",coefficient_F)

coefficient_M= correlation(filtered_m,"reading_score",filtered_m,"writing_score")
print("CoefficientM: ",coefficient_M)
```
