### Plot a pie chart of how many movies of each genre there are 
---


```python
def pie_chart():
  movie_by_genre=df3['genre'].value_counts()
  tot_movies=df3["genre"].count()
  perc_by_genre=movie_by_genre/tot_movies*100
 # print(movie_by_genre)
 # print(perc_by_genre)
# extract your labels using the keys() function
  labels=movie_by_genre.keys()
# plot your pie chart using plt.pie(data, labels = labels)
  plt.pie(movie_by_genre, labels = labels,autopct='%1.1f%%', pctdistance=1.5, labeldistance=1.7)
  plt.show
pie_chart()
```

### Plot a graph showing the mean Rank for each year 

```python
def mean_rk_yr():
  rk_yr=df3.groupby(["year"])["rank"].mean()
#  print (rk_yr)
  labels=rk_yr.keys()
  print(labels)
#print(labels,rk)
# plot the bar chart
  plt.xlabel('Years')#label the axis
  plt.ylabel('Rank')

  plt.bar(labels,rk_yr)
  plt.xticks(rotation=90)
  plt.show()
mean_rk_yr()

```
