````python
import pandas as pd
url = "https://github.com/lilaceri/Working-with-data-/blob/342abab10d93c4bf23b5c55a50f189f12a137c5f/Data%20Sets%20for%20code%20divisio/Covid%20Vaccination%20Data.xlsx?raw=true"
df_Cov_vax=pd.read_excel(url,sheet_name="by_country")
print(df_Cov_vax.info()) 

def clean_df(df,column): 
  df2= df.dropna(subset=[column])
  return df2

df_Cov_vax_cleaned1=clean_df(df_Cov_vax,"total_vaccinations")
print("The first df cleaned has",len(df_Cov_vax_cleaned1),"rows")
modified_df=clean_df(df_Cov_vax_cleaned1,"daily_vaccinations_per_million")
print("The second df cleaned has",len(modified_df),"rows")

def find_median(df,col):
  Median_daily_vax=df[col].median()
  return Median_daily_vax

median=find_median(modified_df,"daily_vaccinations_per_million")
print("The median of daily vaccinations per million is",median)

````

### Challenge 2 - encode daily vaccinations

Write a function to encode daily vaccinations per million, where values greater than or equal to median = 1 and values less than median = 0

````python
#This function creates a column of 1 or 0 based on the values in the kwds
def encode_daily(df, **kwds):
  median=kwds["median"]
  key=kwds["column"]
  if df[key] >= median:
    return 1
  else:
    return 0

modified_df["encoded_daily_vaccinations_per_million"]=modified_df.apply(encode_daily,axis=1,column="daily_vaccinations_per_million",median=median)
print(modified_df["encoded_daily_vaccinations_per_million"].describe())
````

````python
import numpy as np

# Task 1: find the minimum total vaccinations for the United Kingdom
def find_min_vax_country(df, col, country):
  min_vax=df.groupby("country")[col].min()
  min_vax_country=min_vax[country]
  return min_vax_country

min_vax_uk=find_min_vax_country(modified_df,"total_vaccinations","United Kingdom")
minimum_UK_int=int(min_vax_uk) #Save as integer
print(min_vax_uk)
print(minimum_UK_int)

modified_df["To_encode_total_vaccinations"]=modified_df["total_vaccinations"] 
# Task 3: write a function to encode total_vaccinations column 
def encode(df,**kwds):
  t=kwds["threshold"]
  col=kwds["column"]
  if df[col] >= t:
    return 1
  else:
    return 0

modified_df["encoded_total_vaccinations"]=modified_df.apply(encode,axis=1,column="To_encode_total_vaccinations",threshold=minimum_UK_int)
print(modified_df["encoded_total_vaccinations"])

# Task 4: display the countries which total vaccinated is at the same rate or more than the UK
print(modified_df[modified_df["encoded_total_vaccinations"]>0]["country"].unique())

#Task 4: Alternative with numpy
#print(np.unique(modified_df[modified_df["encoded_total_vaccinations"]>0]["country"]))


````
